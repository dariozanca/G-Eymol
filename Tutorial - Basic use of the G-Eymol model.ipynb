# G-Eymol
## Gravitational model of human visual scanpath
See publication here: https://ieeexplore.ieee.org/stamp/stamp.jsp?arnumber=8730418


### Tutorial for a very basic use

```python
import eymol
```

#### Upload an image

```python
import cv2
img = cv2.imread('dummy data/Lenna.png')
```

#### Get raw (simulated) gaze data

```python
gaze_positions = eymol.compute_simulated_scanpath(img, seconds=5) # it may take a while
```

```python
print("Eye positions (Frames per second: 25):\n\n", gaze_positions[:,:2])
```

#### Extract fixations from raw gaze positions

```python
list_of_fixations = eymol.get_fixations(gaze_positions)
```

```python
print("List of fixation points ([pixel row, pixel column, intial time, final time]):\n\n", list_of_fixations)
```
